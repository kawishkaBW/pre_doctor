```python
import missingno as msno
msno.bar(data,color='blue', figsize=[7,5] ,fontsize=10)
```

```python
##Fill NaN value with median value
```

```python
def median_target(var):
    temp=data[data[var].notnull()]
    temp=temp[[var,'target']].groupby(['target'])[var].median().reset_index()
    return temp
```

```python
columns=data.columns
columns=columns.drop('target')

for i in columns:
    median_target(i)
    data.loc[(data['target'] == 0 ) & (data[i].isnull()), i] = median_target(i)[i][0]
    data.loc[(data['target'] == 1 ) & (data[i].isnull()), i] = median_target(i)[i][1]
```

```python
#Detect outliers
#IQR
```

IQR is Q 3 − Q 1 

low outliers are below Q 1 − 1.5 ⋅ IQR ‍   -->    lower limit = Q 1 − 1.5 ⋅ IQR

high outliers are above Q 3 + 1.5 ⋅ IQR ‍  -->    upper limit = Q 3 + 1.5 ⋅ IQR

```python
for feature in data:
    Q1= data[feature].quantile(0.25)
    Q3= data[feature].quantile(0.75)
    IQR = Q3- Q1
    lower = Q1-1.5*IQR
    upper = Q3+1.5*IQR

    if data[(data[feature]>upper)].any(axis=None):
        print(feature +"-","yes")
    else:
        print(feature +"-" ,"No")
```
